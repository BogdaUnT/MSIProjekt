```python
from params.general_params import *
from subsystems.simulator import TtsPidSim
from common.utils import *

system = TtsPidSim(tp, pr)

# Ustawienie parametrów domyślnych
set_default_params(system)

# Ustawienie parametrów charakterystycznych
tend = 15000
system.in_var_val('SPmode', 2)

# Przechowywanie wartości t, SP i PV
t = []
SP = []
PV = []
CV = []
e = []
kp = []
for i in range(int(tend / tp) + 1):
    tk = system.out_var_val('t')
    t.append(tk)
    # Odczyt wartości SP
    sp = system.out_var_val('SP')
    SP.append(sp)
    # Ustawienie wartości kp
    if sp < 0.1 or sp > 0.3:
        k = 5
    else:
        k = 10
    system.in_var_val('kp', k)
    kp.append(k)
    # Krok symulacji
    rval = system.simulation_step(True)
    # Odczyt PV i e
    PV.append(system.out_var_val('PV'))
    e.append(system.out_var_val('e'))
    CV.append(system.out_var_val('CV'))
    progress_bar(tk, tend, 100)

show_evaluation(t, SP, PV, e, CV, None, None, [0.1, 0.3])

plt.figure(figsize=(9, 5))
plt.plot(t, kp, label='kp', color='g')
plt.legend()
```
